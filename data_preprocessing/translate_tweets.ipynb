```python
import pandas as pd
from deep_translator import GoogleTranslator
```

```python
ds = pd.read_csv("../vibe_coding_all_combined.csv")
```

```python
ds.head()
```

```python
# GET DISTRIBUTION OF LANGUAGES IN THE DATASET
ds["lang"].value_counts()

#make it graph
import matplotlib.pyplot as plt
ds["lang"].value_counts().plot(kind="bar")
plt.xlabel("Language")
plt.ylabel("Count")
plt.title("Distribution of Languages in the Dataset")
plt.show()
```

```python
translator = GoogleTranslator(source='auto', target='en')
```

```python
ds['full_text_translated'] = ds.apply(lambda row: translator.translate(row['full_text']) if row['lang'] != 'en' else row['full_text'], axis=1)
```

```python
#save new ds
ds.to_csv("/content/vibe_coding_combined_translated.csv", index=False)
```
