```python
import pandas as pd
import random
import os
import requests
from PIL import Image
from urllib.parse import urlparse
import hashlib
```

```python
df = pd.read_csv("/content/vibe_coding_combined_translated.csv")
df = df[["image_url", "id_str"]]
df
```

```python
image_folder = 'images'
os.makedirs(image_folder, exist_ok=True)
```

```python
print(f"Starting image download to folder: {image_folder}")

success_count = 0
failed_count = 0

for index, row in df.iterrows():
    image_url = row['image_url']
    id_str = row['id_str']

    if pd.isna(image_url):
        # print(f"Skipping row {index}: image_url is NaN for id_str {id_str}")
        continue

    try:
        # Extract file extension from the URL
        parsed_url = urlparse(image_url)
        path = parsed_url.path
        ext = os.path.splitext(path)[1]
        if not ext: # If no extension, try to infer or default to .jpg
            ext = '.jpg'

        filename = f"{id_str}{ext}"
        filepath = os.path.join(image_folder, filename)

        # Check if the file already exists to avoid re-downloading
        if os.path.exists(filepath):
            print(f"Skipping id_str {id_str}: Image already exists at {filepath}")
            success_count += 1
            continue

        response = requests.get(image_url, stream=True, timeout=10)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)

        with open(filepath, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        success_count += 1

    except requests.exceptions.Timeout:
        print(f"Timeout downloading {image_url} for id_str {id_str}")
        failed_count += 1
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {image_url} for id_str {id_str}: {e}")
        failed_count += 1
    except Exception as e:
        print(f"An unexpected error occurred for {image_url} (id_str {id_str}): {e}")
        failed_count += 1

    if (index + 1) % 100 == 0:
        print(f"Processed {index + 1}/{len(df)} images. Successful: {success_count}, Failed: {failed_count}")

print(f"Image download complete. Total successful: {success_count}, Total failed: {failed_count}")
```
